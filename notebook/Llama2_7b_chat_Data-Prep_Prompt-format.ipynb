Data Cleaning for lla-ma2 dataset (Cover Letter)

```python
import pandas as pd
```

```python
# Load data from an Excel workbook
excel_file_path = 'C:/Users/kwanyick/Desktop/NUS-ISS/EBA5004/Prac_Module/Data_Clean/cover_letter_data.xlsx'
df = pd.read_excel(excel_file_path)
df.head()
```

```python

# Select the desired columns
selected_columns = df[[
    'Job Title',
    'Preferred Qualifications',
    'Hiring Company',
    'Applicant Name',
    'Past Working Experience',
    'Current Working Experience',
    'Skillsets',
    'Qualifications'
]]

# Add a new column "System Prompt" with the specified content
df['System Prompt'] = 'Generate a structured cover letter for given user input'

# Create the 'User Prompt' column by concatenating the selected columns
df['User Prompt'] = selected_columns.apply(lambda row: (
    f"Job Title: {row['Job Title']}, "
    f"Preferred Qualifications: {row['Preferred Qualifications']}, "
    f"Hiring Company: {row['Hiring Company']}, "
    f"Applicant Name: {row['Applicant Name']}, "
    f"Past Work Experience: {row['Past Working Experience']}, "
    f"Current Working Experience: {row['Current Working Experience']}, "
    f"Skillsets: {row['Skillsets']}, "
    f"Qualifications: {row['Qualifications']}"
), axis=1)

# Copy the "Cover Letter" column and rename it to "Model Answer"
df['Model Answer'] = df['Cover Letter']


# Combine the selected columns' content into the "text" column with the specified format
df['text'] = (
    '<s>[INST] <<SYS>>\n' + df['System Prompt'] + '\n<</SYS>>\n' +
    df['User Prompt'] + ' [/INST] ' + df['Model Answer'] + ' </s>'
)
```

```python
# Specify the file path to save the output CSV
output_excel_file_path = 'C:/Users/kwanyick/Desktop/NUS-ISS/EBA5004/Prac_Module/Data_Clean/output_file-5.xlsx'

# Save the updated DataFrame to the specified CSV file
df.to_excel(output_excel_file_path, index=False)
```
