Import Libraries and login to Huggingface.

```python
from huggingface_hub import login
from datasets import Dataset
login()
```

Import CSV File using pandas library

```python
pip install pandas
```

```python
import pandas as pd
file_path = "Cover_Letter_Prompt-llama2_text.csv"
```

```python
try:
    input_data = pd.read_csv(file_path, encoding="utf-8")
    # Now you can work with 'input_data'
except FileNotFoundError:
    print(f"File not found: {file_path}")
except Exception as e:
    print(f"An error occurred: {str(e)}")
```

```python
input = input_data
```

Create dataset object from Pandas dataframe, Split dataframe into test and train

```python
dataset = Dataset.from_pandas(input)
dataset  = dataset.train_test_split(test_size=0.3)

# Print the dataset
print(dataset)
```

Go to your hugginface profile > New dataset and create new dataset with above screen.

Once dataset repository has been created , run below command to upload dataset on huggingface.

```python
dataset.push_to_hub("kwanyick/cover-letter-dataset-text-prompt")
```
